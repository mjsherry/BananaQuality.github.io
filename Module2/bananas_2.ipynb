```python
import pandas as pd
data = pd.read_csv('/Users/mikaelasherry/790/Module2/banana_quality.csv')
```

```python
data = data.head(4000)

data
```

```python
import altair as alt
```

```python
alt.Chart(data).mark_bar().encode(
    x = alt.X('HarvestTime', title='Harvest Time'),
    y = "Sweetness",
    color = alt.Color('Quality', scale=alt.Scale(domain=['Bad', 'Good'], range=['red', 'green']))
).properties(
    title='Harvest Time vs. Sweetness'
    
)
```

```python
alt.Chart(data).mark_circle().encode(
    x = "HarvestTime",
    y = "Sweetness",
    color="Quality"
)
```

```python
alt.Chart(data).mark_circle().encode(
    x = "HarvestTime",
    y = "Ripeness",
    color="Quality"
)
```
